```python
import matplotlib.pyplot as plt
import re
```

```python
first_fit_malloc_time = []
best_fit_malloc_time = []

first_fit_loop_time = []
best_fit_loop_time = []

first_fit_util_before_free = []
best_fit_util_before_free = []

first_fit_util_after_free = []
best_fit_util_after_free = []

pattern = r"\d+\.?\d*"

with open("first-fit.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        match = re.findall(pattern, line)
        if line.startswith("time of malloc()"):
            first_fit_malloc_time.append(int(match[0]))
        elif line.startswith("before free"):
            first_fit_util_before_free.append(float(match[0]))
            first_fit_util_after_free.append(float(match[1]))
        elif line.startswith("time of loop"):
            first_fit_loop_time.append(int(match[1]))

with open("best-fit.txt", "r") as file:
    lines = file.readlines()
    for line in lines:
        match = re.findall(pattern, line)
        if line.startswith("time of malloc()"):
            best_fit_malloc_time.append(int(match[0]))
        elif line.startswith("before free"):
            best_fit_util_before_free.append(float(match[0]))
            best_fit_util_after_free.append(float(match[1]))
        elif line.startswith("time of loop"):
            best_fit_loop_time.append(int(match[1]))

print(first_fit_malloc_time)
print(best_fit_malloc_time)
print(first_fit_loop_time)
print(best_fit_loop_time)
print(first_fit_util_before_free)
print(best_fit_util_before_free)
print(first_fit_util_after_free)
print(best_fit_util_after_free)
```

```python
plt.plot(first_fit_malloc_time, label="first-fit")
plt.plot(best_fit_malloc_time, label="best-fit")
plt.xlabel("Loop")
plt.ylabel("Malloc time (ms)")
plt.title("Comparison of first-fit and best-fit in malloc time")
plt.xticks(range(20))
plt.legend()
plt.show()
```

```python
plt.plot(first_fit_loop_time, label="first-fit")
plt.plot(best_fit_loop_time, label="best-fit")
plt.xlabel("Loop")
plt.ylabel("Loop time (ms)")
plt.title("Comparison of first-fit and best-fit in loop time")
plt.xticks(range(20))
plt.legend()
plt.show()
```

```python
plt.plot(first_fit_util_before_free, label="first-fit")
plt.plot(best_fit_util_before_free, label="best-fit")
plt.xlabel("Loop")
plt.ylabel("Util before free")
plt.title("Comparison of first-fit and best-fit in util before free")
plt.xticks(range(20))
plt.legend()
plt.show()
```

```python
plt.plot(first_fit_util_after_free, label="first-fit")
plt.plot(best_fit_util_after_free, label="best-fit")
plt.xlabel("Loop")
plt.ylabel("Util after free")
plt.title("Comparison of first-fit and best-fit in util after free")
plt.xticks(range(20))
plt.legend()
plt.show()
```
